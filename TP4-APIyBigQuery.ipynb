```python
import requests
import pandas as pd

diccionario = {"fecha":[],"ofertaMax":[],"ofertaMin":[], "nombreSubasta":[]}
respuesta = requests.get('https://whiskyhunter.net/api/auctions_data/')
#limit response to 50
respuesta = respuesta.json()[:50]

print(respuesta)
print("--------------------------------------------------")

for renglon in respuesta:
    diccionario["fecha"].append(renglon["dt"])
    diccionario["ofertaMax"].append(renglon["winning_bid_max"])
    diccionario["ofertaMin"].append(renglon["winning_bid_min"])
    diccionario["nombreSubasta"].append(renglon["auction_name"])

print(diccionario)

data = pd.DataFrame(diccionario)
data.to_csv("./datosSubasta.csv")

```

```python


import requests
import pandas as pd

diccionario = {"nombre":[],"pais":[],"rating":[]}
respuesta = requests.get('https://whiskyhunter.net/api/distilleries_info/')
#limit response to 50
respuesta = respuesta.json()[:50]

print(respuesta)
print("--------------------------------------------------")

for renglon in respuesta:
    diccionario["nombre"].append(renglon["name"])
    diccionario["pais"].append(renglon["country"])
    diccionario["rating"].append(renglon["whiskybase_rating"])


print(diccionario)

data = pd.DataFrame(diccionario)
data.to_csv("./datosDestilerias.csv")
```
